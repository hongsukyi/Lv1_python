## 16.	산점도 Scatter Plot 

일반적으로 사용되는 또 다른 플롯의 유형은 간단한 산점도입니다. 개별적인 점 혹은 원이나 다양한 모양으로 데이터를 표시합니다

```python
x7 = np.linspace(0, 10, 80)
y7 = np.sin(x7)
plt.plot(x7, y7, 'o', color = 'black');
```

위의 코드에서 함수 호출 시 세 번째 인수는 플로팅에 사용할 기호 유형을 지정하는 문자입니다. 선 스타일을 제어할 때는 -와 -- 같은 옵션을 지정하고, 마커 스타일에는 고유한 짧은 문자열 코드 집합을 사용합니다. 사용 가능한 기호의 전체 목록은 plt.plot 또는 Matplotlib의 온라인 매뉴얼을 찾아보십시오. 대부분의 기호는 상당히 직관적입니다. 아래에서 일반적으로 사용되는 것들을 나열하였으니 참고하세요.

```python
rng = np.random.RandomState(0)
fig = plt.figure(figsize=(10,8))
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']: 
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
    plt.legend(numpoints=1)
    plt.xlim(0, 1.8);
```

점을 연결하는 선과 함께 그리기 위해 선 및 색상 코드와 함께 사용할 수 있습니다.

- plt.plot과의 주요 차이점은 크기, 면색, 가장자리 색 등 점의 속성을 개별적으로 제어하거나,

- 데이터에 매핑할 수 있는 산점도를 만드는 데 사용할 수 있다는 것입니다. 

- 다양한 색상과 크기의 점으로 임의의 산점도를 작성. 

- 겹치는 결과를 더 잘 보려면 alpha 키워드를 사용하여 투명도 레벨을 조정할 수 있습니다.

```python
rng = np.random.RandomState(0)
x = rng.randn(100); y = rng.randn(100)
colors = rng.rand(100); sizes = 1000 * rng.rand(100)
plt.scatter(x, y, c=colors,s=sizes, alpha=0.3,cmap='viridis')
plt.colorbar();
```

## 17.	Histogram

히스토그램은 데이터셋을 이해하는 데 도움이 될 수 있습니다. 아래는 히스토그램 함수를 간단히 사용한 예입니다.

```python
data1 = np.random.randn(1000)
plt.hist(data1); 
plt.hist(data1, bins=30, density=True, alpha=0.5, histtype='stepfilled', color='steelblue',
         edgecolor='none');
```

plt.hist의 온라인 매뉴얼에는 다양한 옵션에 대한 더욱 자세한 정보가 있으니 참고하세요.

histitype = 'stepfilled' 옵션에 alpha값 지정을 통해 투명도를 조절하면 다양한 분포에 대한 히스토그램을 비교할 때 매우 유용합니다.

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.3, density=True, bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

# 35. Seaborn을 이용한 시각화

Matplotlib은 유용하고 많은 사용자를 보유한 시각화 도구로 입증되었지만 아쉬운 점은 있다. 
종종 발생하는 Matplotlib에 대한 몇 가지 유효한 불만 사항: 

- 버전 2.0 이전의 Matplotlib은 기본 옵션값을 선택했을 때 디자인이 만족스럽지 않다. 이는 1999년 경 MATLAB을 대체하고자 만들어진 것으로 그 당시 수준에 머물러 있는 요소들이 아직 남아 있습니다.

- Matplotlib의 API는 상대적으로 수준이 낮습니다. 정교한 통계적 시각화가 가능한데 막상 구현하려면 상용구 코드가 많이 필요합니다.

- Matplotlib는 Pandas보다 10년 이상 앞서 있었기 때문에 Pandas DataFrame과 함께 사용하도록 설계되지 않았습니다. Pandas DataFrame에서 데이터를 시각화하려면 각 Series를 추출해야하며 올바른 형식으로 변환해야 합니다. 플롯에서 DataFrame 레이블을 지능적으로 사용할 수 있는 플로팅 라이브러리가 있으면 더 좋을 것입니다.

위의 단점을 보완할 해결책은 Seaborn입니다. Seaborn은 Matplotlib 위의 API로서, 현 시점에서 적절한 플롯 스타일 및 색상 기본값을 가지고 있고, 일반적인 통계 플롯 유형에 대한 수준 높은 기능을 간단히 구현할 수 있으며, Pandas DataFrame과 직접 연계할 수 있습니다.

## plt.subplot : 서브 플롯의 그리드



```python
for i in range(1, 7):
    plt.subplot(2, 3, i) 
    plt.text(0.5, 0.5, str((2, 3, i)),
             fontsize=18, ha='center')
```

plt.subplots_adjust 명령어는 이러한 플롯 간의 간격을 조정하는 데 사용할 수 있습니다. 다음의 코드는 위의 코드와 작동 방식이 동등한 객체 지향 명령어인 fig.add_subplot()을 사용한 것입니다.

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4)
for i in range(1, 7): 
    ax = fig.add_subplot(2, 3, i)
    ax.text(0.5, 0.5, str((2, 3, i)),
            fontsize=18, ha='center')
```

그림의 높이와 너비에 따른 간격을 서브플롯 크기 단위로 지정하는 plt.subplots_adjust 명령의 hspace 및 wspace 인수를 사용했습니다. 위의 경우 간격은 서브플롯 너비와 높이의 40%로 설정한 것입니다.

## plt.GridSpec : 보다 복잡한 배치

규칙적으로 배치하는 그리드 대신 여러 행과 열을 가로지르는 서브플롯을 생성하려면 plt.GridSpec()이 가장 좋은 도구입니다. plt.GridSpec() 객체는 자체적으로 플롯을 작성하지 않습니다. 이 객체는 단순히 plt.subplot() 명령어가 인식하는 편리한 인터페이스일 뿐입니다. 예를 들어, 폭 및 높이 여백이 지정된 두 개의 행과 세 개의 열로 이루어진 그리드에 대한 GridSpec은 다음과 같습니다.

```python
grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.3)

plt.subplot(grid[0, 0])
plt.subplot(grid[0, 1:])
plt.subplot(grid[1, :2])
plt.subplot(grid[1, 2]);
```

```python
이러한 유형의 유연한 그리드 정렬은 다양한 용도로 사용됩니다. 아래와 같이 다축 히스토그램 플롯을 만들 때도 자주 사용합니다.
```

```python
mean = [0, 0] 
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 3000).T

fig = plt.figure(figsize=(6, 6))
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

main_ax.plot(x, y, 'ok', markersize=3, alpha=0.2)
x_hist.hist(x, 40, histtype='stepfilled',orientation='vertical', color='gray')
x_hist.invert_yaxis()
y_hist.hist(y, 40, histtype='stepfilled',orientation='horizontal', color='gray')
y_hist.invert_xaxis()
```
